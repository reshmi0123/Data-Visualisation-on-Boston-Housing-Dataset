```python
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 


from sklearn.datasets import load_boston
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target
boston.head()
```

```python
boston.info()
```

```python
boston.describe()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(boston['MEDV'], bins=30)
plt.show()

##Distribution of the target variable MEDV

## We see that the values of MEDV are distributed normally with few outliers.
```

```python
sns.relplot(x = 'DIS', y = 'AGE', data = boston, size='AGE',hue='AGE')

plt.show()

##Correlation between age of the house and distance to Boston Employment Centre

## Scatter plot show us that most of the old houses are closer to the Boston Employment Centre.
```

```python
sns.relplot(x = 'NOX', y = 'DIS', data = boston, size='NOX',hue='NOX')

##Correlation between Pollution level among houses and distance to Boston Employment Centre

##We found that the pollution level are high among old houses closer to the Boston employment centre.
```

```python
sns.relplot(x = 'CRIM', y = 'AGE', data = boston, size='CRIM',hue='CRIM')

##Correlation between Crime Rate and Age of the house 

##We found that the per capita crime rate mainly concentrated among old houses closer to the Boston employment centre.
```

```python
sns.relplot(x = 'DIS', y = 'LSTAT', data = boston,hue='CRIM',size='CRIM', col='CHAS')

plt.show()

##LSTAT (Percentage of lower status of the population) Vs Distance factor being,Crime Rate Tageting CHAS var

##In the above graph shows that per capita crime rate is correlated with areas away from the Charles River.
```

```python
correlation_matrix = boston.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)

##Correlation matrix measures the linear relationships between the variables.

##The correlation coefficient ranges from -1 to 1. 
# If the value is close to 1, it means that there is a strong positive correlation between the two variables. 
# When it is close to -1, the variables have a strong negative correlation.
```
